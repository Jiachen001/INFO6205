```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
orderedData=pd.read_csv("orderedInput.csv")
partialOrderedData=pd.read_csv("particialOrderedInput.csv")
randomData=pd.read_csv("randomInput.csv")
reversedData=pd.read_csv("reversedInput.csv")
```

```
plt.figure(figsize=(15,10))
plt.plot(np.log(orderedData["N"]),np.log(orderedData["Time"]),label="orderedData")
plt.plot(np.log(partialOrderedData["N"]),np.log(partialOrderedData["Time"]),label="partialOrderedData")
plt.plot(np.log(randomData["N"]),np.log(randomData["Time"]),label="randomData")
plt.plot(np.log(reversedData["N"]),np.log(reversedData["Time"]),label="reversedData")
plt.xlabel("log(N)")
plt.ylabel("log(Time)")
plt.legend()
```
