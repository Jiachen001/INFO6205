```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data1=pd.read_csv("depth2loops.csv")
data2=pd.read_csv("size1loop.csv")
data3=pd.read_csv("size2loops.csv")

plt.plot(data1['N'],data1['Time'],label="depth2loops")
plt.plot(data2['N'],data2['Time'],label="size1loop")
plt.plot(data3['N'],data3['Time'],label="size2loop")
plt.xlabel("N")
plt.ylabel("Time")
plt.legend();

```
